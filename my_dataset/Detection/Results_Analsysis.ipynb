<h2>My Dataset Acuracy Analsis on Epochs</h2>

```python
import numpy as np
import matplotlib.pyplot as plt

# Accuracy
# Performance measures for each model and class
models = ['CNN', 'MLP', 'LSTM', 'CNN-LSTM', 'Bi-LSTM', 'CNN-Bi-LSTM']
epochs_20 = [98.88, 67.46, 73.29, 98.87, 99.04, 98.77]
epochs_30 = [99.01, 71.67, 73.72, 98.99, 99.08, 98.97]
epochs_50 = [98.98, 73.04, 74.61, 99.05, 99.20, 99.13]
epochs_100 = [99.07, 74.92, 74.53, 99.25, 99.34, 99.40]
epochs_150 = [99.08, 77.51, 74.83, 99.30, 99.36, 99.50]

# Set the width of the bars
bar_width = 0.10

# Set the positions of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + 2*bar_width for x in r1]
r4 = [x + 3*bar_width for x in r1]
r5 = [x + 4*bar_width for x in r1]

# Create the bar chart with a white background and a grid
plt.figure(figsize=(8, 4))
plt.gca().set_facecolor('white')  # Set background color to white

# Create the bars
plt.bar(r1, epochs_20, color='b', width=bar_width, edgecolor='grey', label='20 Epochs')
plt.bar(r2, epochs_30, color='r', width=bar_width, edgecolor='grey', label='30 Epochs')
plt.bar(r3, epochs_50, color='y', width=bar_width, edgecolor='grey', label='50 Epochs')
plt.bar(r4, epochs_100, color='g', width=bar_width, edgecolor='grey', label='100 Epochs')
plt.bar(r5, epochs_150, color='m', width=bar_width, edgecolor='grey', label='150 Epochs')

# Add vertical lines to separate each model's bars
for i in range(len(models) - 1):
    plt.axvline(x=r5[i] + bar_width / 2, color='gray', linewidth=0.3)

# Add x-axis labels and tick labels
plt.xlabel('Implemented Deep Learning Models', fontsize=12)
plt.ylabel('Accuracy (0% to 100%)', fontsize=12)
#plt.title('Model Accuracy on different Epochs\n', fontsize=14)

# Customize x-axis tick labels
plt.xticks([r + 2*bar_width for r in range(len(r1))], models, fontsize=12)

# Place the legend at the bottom left corner of the plot
plt.legend(loc='lower left', fontsize=10, ncol=1)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Performance measures for each model and class
models = ['CNN', 'MLP', 'LSTM', 'CNN-LSTM', 'Bi-LSTM', 'CNN-Bi-LSTM']
epochs_20 = [0.1629, 0.9110, 0.5163, 0.0383, 0.0318, 0.0435]
epochs_30 = [0.0863, 0.7781, 0.5040, 0.0332, 0.0265, 0.0343]
epochs_50 = [0.0792, 0.9123, 0.4983, 0.0305, 0.0193, 0.0285]
epochs_100 = [0.0444, 0.8184, 0.4719, 0.0171, 0.0124, 0.0185]
epochs_150 = [0.0412, 0.7559, 0.4678, 0.0126, 0.0095, 0.0139]

# Set the width of the bars
bar_width = 0.10

# Set the positions of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width + 0.05 for x in r1]
r3 = [x + 2*bar_width + 0.1 for x in r1]
r4 = [x + 3*bar_width + 0.15 for x in r1]
r5 = [x + 4*bar_width + 0.2 for x in r1]

# Create the bar chart with a white background
plt.figure(figsize=(8, 4))
plt.gca().set_facecolor('white')

# Create the bars
plt.bar(r1, epochs_20, color='b', width=bar_width, edgecolor='grey', label='20 Epochs')
plt.bar(r2, epochs_30, color='r', width=bar_width, edgecolor='grey', label='30 Epochs')
plt.bar(r3, epochs_50, color='y', width=bar_width, edgecolor='grey', label='50 Epochs')
plt.bar(r4, epochs_100, color='g', width=bar_width, edgecolor='grey', label='100 Epochs')
plt.bar(r5, epochs_150, color='m', width=bar_width, edgecolor='grey', label='150 Epochs')

# Add vertical lines to separate each model's bars
for i in range(len(models) - 1):
    plt.axvline(x=r5[i] + bar_width / 2, color='gray', linewidth=0.5)

# Add x-axis labels and tick labels
plt.xlabel('Deep Learning Models', fontsize=12)
plt.ylabel('Loss', fontsize=12)
#plt.title('Model Training and Testing Loss on Epochs\n', fontsize=16)

# Customize x-axis tick labels
plt.xticks(r1 + 2.5*bar_width, models, fontsize=12)

# Place a legend on the plot at the top left
plt.legend(loc='upper left', bbox_to_anchor=(0, 1), ncol=1, fontsize=10)

# Set the limits of the y-axis to accommodate the bars
max_loss = max(max(epochs_20), max(epochs_30), max(epochs_50), max(epochs_100), max(epochs_150))
plt.ylim(0, max_loss + max_loss * 0.1)

# Display the chart
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust the rect to make room for the legend
plt.show()
```
