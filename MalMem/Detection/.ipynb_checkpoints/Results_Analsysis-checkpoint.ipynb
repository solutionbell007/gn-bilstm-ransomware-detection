<h2>Acuracy Analsis on Epochs</h2>

```python
import numpy as np
import matplotlib.pyplot as plt

# Performance measures for each model and class
models = ['CNN', 'DBN', 'ELM', 'MLP', 'LSTM', 'Bi-LSTM', 'CNN-LSTM', 'CNN-Bi-LSTM']
epochs_20 = [99.90, 0.0003, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005]
epochs_30 = [99.92, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005]
epochs_50 = [99.93, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005]
epochs_70 = [99.96, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005]
epochs_100 = [22.5, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005, 0.0005]

# Set the width of the bars
bar_width = 0.15

# Set the positions of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width + 0.05 for x in r1]
r3 = [x + 2*bar_width + 0.1 for x in r1]
r4 = [x + 3*bar_width + 0.15 for x in r1]
r5 = [x + 4*bar_width + 0.2 for x in r1]  # Adjust the offset here

# Create the bar chart with a white background and no grid
plt.figure(figsize=(15, 5))
plt.gca().set_facecolor('white')  # Set background color to white
plt.grid(False)  # Remove grid

plt.bar(r1, epochs_20, color='b', width=bar_width, edgecolor='white', linewidth=2.2, label='Accuracy on 20 Epochs')
plt.bar(r2, epochs_30, width=bar_width, color='r', edgecolor='white', linewidth=2.2, label='Accuracy on 30 Epochs')
plt.bar(r3, epochs_50, color='y', width=bar_width, edgecolor='white', linewidth=2.2, label='Accuracy on 50 Epochs')
plt.bar(r4, epochs_70, color='g', width=bar_width, edgecolor='white', linewidth=2.2, label='Accuracy on 70 Epochs')
plt.bar(r5, epochs_100, color='m', width=bar_width, edgecolor='white', linewidth=2.2, label='Accuracy on 100 Epochs')

# Add x-axis labels and tick labels
plt.xlabel('')
plt.ylabel('')
plt.title('Model Training and Testing Accuracy on Epochs\n')

# Customize x-axis tick labels
plt.xticks(r1 + 2.5*bar_width + 0.025, models)  # Adjust the offset here

# Add a legend as a horizontal line below the plot
legend = plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=4, handlelength=1, handletextpad=0.5)

# Set the size of the bars in the legend to make them square
for handle in legend.legendHandles:
    handle.set_height(10)  # Adjust the size here
    handle.set_width(10)  # Adjust the size here

# Display the chart
plt.tight_layout()
plt.show()
```
